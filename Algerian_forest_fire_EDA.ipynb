```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleanup

### Basic EDA

#### missing values

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
```

```python
dataset.isnull().sum()
```

#### Removing null values

```python
df=dataset.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

#### drop row with index 122

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.head()
```

```python
df.columns
```

#### Fix spaces in column names

```python
df.columns=df.columns.str.strip()
df.columns
```

#### change the required columns to int

```python
df[['day','month','year','RH','WS','Temperature']]=df[['day','month','year','RH','Ws','Temperature']].astype(int)
```

```python
df.head()
```

#### changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtype=='O']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

### Advanced EDA

```python
## Save the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_CLEANED.csv',index=False)
```

```python
df.head()
```

#### Drop the non required columns

```python
df_copy=df.drop(['month','year','day'],axis=1)
```

```python
df_copy.head()
```

#### Encoding of categories in classes

```python
df_copy['Classes']=np.where(df_copy['Classes']=='not fire',0,1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
df_copy['Classes'].value_counts()
```

#### Plot density plot for all features

```python
plt.style.use('dark_background')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
##percentage pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
##plotting pie chart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Percentage of fire and not fire")
plt.show()
```
